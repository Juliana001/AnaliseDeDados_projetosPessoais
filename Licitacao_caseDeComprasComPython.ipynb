#Importando bibliotecas

```python
import os
import pandas as pd
```

```python
arquivos = os.listdir()
#Lê os dados
licitacao = pd.read_csv('empresa_A.csv')

#Altera o nome de uma coluna
licitacao = licitacao.rename(columns={'Total (R$)' : 'Empresa A'})

#Exclui uma coluna
licitacao = licitacao.drop(['Valor unitário (R$)'], axis=1)

empresas = arquivos[1:5]
for empresa in empresas:
  proposta = pd.read_csv(empresa)
  licitacao[empresa[:-4]] = proposta['Total (R$)'].copy()

licitacao = licitacao.rename(columns={'empresa_D' : 'Empresa D'})
licitacao = licitacao.rename(columns={'empresa_C' : 'Empresa C'})
licitacao = licitacao.rename(columns={'empresa_E' : 'Empresa E'})
licitacao = licitacao.rename(columns={'empresa_B' : 'Empresa B'})
#Mostra os dados em forma de tabela
display(licitacao)
```

#Verificar valores totais

```python
for coluna in licitacao.columns[3:]:
  total = licitacao[coluna].sum()
  print("Empresa {} : R$ {:,.2f}".format(coluna, total))
```

#Resultado ótimo

```python
licitacao['Minimo'] = licitacao.drop(['Item', 'Descrição item', 'Quantidade requerida'], axis = 1).min(axis=1)

display(licitacao)

minimo = licitacao['Minimo'].sum()
print("O valor mínimo é: R$ ", minimo)
```

#Planilha de desvios por fornecedor

```python
for empresa in licitacao.columns[3:]:
  desvio = licitacao[['Item', 'Descrição item', 'Quantidade requerida', empresa, 'Minimo']]
  desvio['Delta'] = desvio[empresa]-desvio['Minimo']
  display(desvio)
```
