###Este projeto teve início com foco em Business Intelligence, voltado à visualização dos dados do ENADE 2023, mas evoluiu para uma abordagem mais abrangente de Ciência de Dados e Engenharia de Dados. Nesta primeira etapa, o objetivo principal foi a limpeza e exploração inicial dos dados. O processo começou com a remoção das linhas que não continham informações sobre as universidades efetivamente avaliadas. Em seguida, foi identificada a presença de espaços nos nomes de algumas colunas, o que dificultava sua manipulação; por esse motivo, todos os nomes de colunas foram padronizados utilizando o estilo camelCase. Durante a análise exploratória, observou-se que um número significativo de universidades participou do processo avaliativo, mas não possuía nota do ENADE. Diante disso, optou-se por dividir os dados em dois datasets: um contendo apenas as instituições com nota ENADE e outro reunindo aquelas sem nota, facilitando análises futuras e garantindo maior clareza na exploração dos dados.

#6 - renomeação das colunas

```python
dados.rename(columns = {
    dados.columns[0]: 'ano',
    dados.columns[1]: 'codigoArea',
    dados.columns[2]: 'areaAvaliacao',
    dados.columns[3]: 'grauAcademico',
    dados.columns[4]: 'codigoIES',
    dados.columns[5]: 'nomeIES',
    dados.columns[6]: 'siglaIES',
    dados.columns[7]: 'organizacaoAcademica',
    dados.columns[8]: 'categoriaAdministrativa',
    dados.columns[9]: 'codigoCurso',
    dados.columns[10]: 'modalidadeEnsino',
    dados.columns[11]: 'codigoMunicipio',
    dados.columns[12]: 'municipioDoCurso',
    dados.columns[13]: 'siglaUF',
    dados.columns[14]: 'qtdConcluintesInscritos',
    dados.columns[15]: 'qtdConcluintesParticipantes',
    dados.columns[16]: 'notaBrutaFG',
    dados.columns[17]: 'notaPadronizadaFG',
    dados.columns[18]: 'notaBrutaCE',
    dados.columns[19]: 'notaPadronizadaCE',
    dados.columns[20]: 'conceitoEnadeContinuo',
    dados.columns[21]: 'conceitoEnadeFaixa',
    dados.columns[22]: 'beneficenteAssistenciaSocialCEBAS',
}, inplace=True)
```
